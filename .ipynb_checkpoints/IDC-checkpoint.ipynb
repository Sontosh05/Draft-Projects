```julia
using Plots
include("polyfit_function.jl")
include("Lagrange_interpolation.jl")
```

```julia
function idc(f, a, b, y0, h, N, p, M, K)
    
    # Inputs: a: tspan[1], b: tspan[2], y0:initial cond, h=dt, N: # of intervals, 
    # Inputs: p= desired order, M : # of subintervals on each group, K: # of corrections,
    
    
    J = N ÷ M # Number of groups (÷ denotes the integer division)
    
   
    
    
    y=zeros(1, M+1, M+1)
    t = zeros(M+1)
    t0 = a
    
    for j = 1:J 
        
        t[1] = t0 # t0 gets updated after every group J
        
        # Prediction loop
        
        y[1, 1, 1] = y0 # Initial condition for each group
        
        for m = 1:M
            t[m+1] = t[1] + m * h # time increment for each group
            y[1, m+1, 1]= y[1, m, 1]+ h* f(t[m], y[1, m, 1]) # Forward Euler update
        end
        
        # Correction loop
        
        for k =1:K
            
            y[1, 1, k+1]= y[1, 1, k]
            
            nfval= f(t, y[1, :, k]) # New function value 
            
            for m=1:M
                
                Q=Lagrange(t[1:k+1], nfval[1:k+1], t[m], t[m+1])  # Quadrature
                y[1, m+1, k+1] = y[1, m, k+1] + h*(f(t[m], y[1, m, k+1]) - f(t[m], y[1, m, k]))+Q
                
            end
        end
        
        y0=y[1, M+1, K+1] # storing initial value for the next group (Predictor)
        
        t0 = t[M+1] # updated initial time
        
        
    end
    
    return y 
    
end
```

```julia
a = 0 # Initial time
b = 2 # Terminal time
y0 = 1 # Initial condition
p = 6 # desired order
M = p - 1 # Number of corrections
K=M
Nk = 10

err = zeros(p, Nk)
ns = zeros(Nk)

for kk in 1:p
    for n in 1:Nk
        N = 50 * n
        ns[n] = N
        h = (b - a) / N
        #f(t, y)= -2* t .* y .^2
        f(t, y)= y
        #f(t, y)= y.^2
        ystore= idc(f, a, b, y0, h, N, p, M, kk - 1)

        #exact = 1.0 / 2
        exact=exp(b)
        #exact=2.0
        err[kk, n] = abs(exact - ystore[1, end, kk])
    end
end
# Compute order of convergence and error plot
pdeg = polyfit(log10.(ns[:]), log10.(err[1, :]),1)
print("Order p= ",abs(pdeg[2]),"\n") 
plot(ns, err[1, :],xaxis=:log, yaxis=:log, label="Predictor")

for j in 2:p
    plot!(ns, err[j, :],xaxis=:log, yaxis=:log, label="$(j-1) Correction")
    pdeg = polyfit(log10.(ns[:]), log10.(err[j, :]),1)
    print("Order p= ", abs(pdeg[2]),"\n")
end
display(plot!(legend=:bottomleft))
```

# High-Order Integrators (RK-2) in IDC

```julia
function idc_highO(f, a, b, y0, h, N, p, M, K)
    
    # Inputs: a: tspan[1], b: tspan[2], y0:initial cond, h=dt, N: # of intervals, 
    # Inputs: p= desired order, M : # of subintervals on each group, K: # of corrections,
    
    
    J = N ÷ M # Number of groups (÷ denotes the integer division)
    
    # Construct a 4 dimensional Array for storing y
    # y=zeros(1, M+1, J, K+1) Here 1: Scalar ODE, M+1: # of Nodes, 
    #   J: Group, and M+1: Predictor + # of Corrections
    
    
    y=zeros(1, M+1, M+1)
    t = zeros(M+1)
    t0 = a
    
    for j = 1:J 
        
        t[1] = t0 # t0 gets updated after every group J
        
        # Prediction loop
        
        y[1, 1, 1] = y0 # Initial condition for each group
        
        for m = 1:M
            t[m+1] = t[1] + m * h # time increment for each group
            y[1, m+1, 1]= y[1, m, 1]+ h* f(t[m], y[1, m, 1]) # Forward Euler update
        end
        
        # Correction loop
        
        for k =1:K
            
            y[1, 1, k+1]= y[1, 1, k]
            
            nfval= f(t, y[1, :, k]) # New function value 
            
            for m=1:M
                
                Q=Lagrange(t[1:k+1], nfval[1:k+1], t[m], t[m+1])  # Quadrature
                y[1, m+1, k+1] = y[1, m, k+1] + h*(f(t[m], y[1, m, k+1]) - f(t[m], y[1, m, k]))+Q
                
            end
        end
        
        y0=y[1, M+1, K+1] # storing initial value for the next group (Predictor)
        
        t0 = t[M+1] # updated initial time
        
        
    end
    
    return y 
    
end
```

# Multistep IDC

```julia
function idc_multistep(f, a, b, y0, h, N, p, M, K)
    
    # Inputs: a: tspan[1], b: tspan[2], y0:initial cond, h=dt, N: # of intervals, 
    # Inputs: p= desired order, M : # of subintervals on each group, K: # of corrections,
    
    
    J = N ÷ M # Number of groups (÷ denotes the integer division)
    
    
    
    
    y=zeros(1, M+1, M+1)
    t = zeros(M+1)
    t0 = a
    
    for j = 1:J 
        
        t[1] = t0 # t0 gets updated after every group J
        
        # Prediction loop
        
        y[1, 1, 1] = y0 # Initial condition for each group
        
        k1=h*f(t[1], y[1, 1, 1])
        k2=h*f(t[1]+h, y[1, 1, 1]+k1)
        y[1, 2, 1]=y[1]+0.5*(k1+k2)
        
        for m = 1:M-1
            t[m+1] = t[1] + m * h # time increment for each group
            y[1, m+2, 1]= y[1, m+1, 1]+ (3/2)*h* f(t[m+1], y[1, m+1, 1]) -0.5*h* f(t[m], y[1, m, 1]) # AB-2 update
        end
        
        # Correction loop
        
        for k =1:K
            
            y[1, 1, k+1]= y[1, 1, k]
            
            nfval= f(t, y[1, :, k]) # New function value 
            
            # First step computed by Trapezoidal RK2 (Explicit)
            Qdr=Lagrange(t[1:(2*k+2)], nfval[1:(2*k+2)], t[1], t[2]) # Quadrature
            
            k1= h*(f(t[1], y[1, 1, k+1])- f(t[1], y[1, 1, k]))
            k2= h*(f(t[1] + h, y[1, 1, k+1] + k1 + Qdr) - f(t[1]+h, y[1, 2, k])) 
        
            y[1, 2, k+1]=y[1, 1, k+1] + k1/2 + k2/2 + Qdr
            
            
            for m=2:M
                
                Q= Lagrange(t[1:(2*k+2)], nfval[1:(2*k+2)], t[m], t[m+1])
                b1= (3/2)* h*(f(t[m], y[1, m, k+1]) - f(t[m], y[1, m, k]))
                b2= (1/2)* h*(f(t[m-1], y[1, m-1, k+1]) - f(t[m-1], y[1, m-1, k]))
                y[1, m+1, k+1]= y[1, m, k+1] + b1 - b2 + Q
                
            end
        end
       
        
        y0=y[1, M+1, K+1] # storing initial value for the next group (Predictor)
        
        t0 = t[M+1] # updated initial time
        
        
    end
    
    return y 
    
end
```

```julia
a = 0 # Initial time
b = 1 # Terminal time
y0 = 1 # Initial condition
p = 6 # desired order
M = p - 1 # Number of corrections

K=M

Nk = 8

err = zeros(p, Nk)
ns = zeros(Nk)

for kk in 1:2
    for n in 1:Nk
        N = 20 * n
        ns[n] = N
        h = (b - a) / N
        
        f(t, y)= y
        
        ystore= idc_multistep(f, a, b, y0, h, N, p, M, kk - 1)

        
        exact=exp(b)
        
        err[kk, n] = abs(exact - ystore[1, end, kk])
    end
end
# Compute order of convergence and error plot
pdeg = polyfit(log10.(ns[:]), log10.(err[1, :]),1)
print("Order p= ",abs(pdeg[2]),"\n") 
plot(ns, err[1, :],xaxis=:log, yaxis=:log, label="Predictor")

for j in 2:2
    plot!(ns, err[j, :],xaxis=:log, yaxis=:log, label="$(j-1) Correction")
    pdeg = polyfit(log10.(ns[:]), log10.(err[j, :]),1)
    print("Order p= ", abs(pdeg[2]),"\n")
end
display(plot!(legend=:bottomleft))
```
