```julia
using LinearAlgebra, DSP, Polynomials

function LagF(t, g, x)
    N = length(t)
    L = zeros(N, N)

    for i in 1:N
        p = [1]
        for j in 1:N
            if i != j
                p = conv(p, [-t[j], 1]) / (t[i] - t[j])
            end
        end
        L[i, :] = p * g[i]
    end

    apprx = sum(L, dims=1)  # Reverse the order of coefficients
    
    poly = Polynomial(apprx[1, :])  # Create a polynomial
    
    
    result = poly(x)  # Evaluate the polynomial at a specific point x
    
    return result
end
```

```julia
function idc6_FE(f, a, b, y0, h, N, p, M, K)
    
    # Inputs: a: tspan[1], b: tspan[2], y0: initial cond, h=dt, N: # of intervals, 
    # Inputs: p= desired order, M : # of subintervals on each group, K: # of corrections,
    
    
    J = N ÷ M # Number of groups (÷ denotes the integer division)
    

    y=zeros(1, M+1, K+1)
    t = zeros(M+1)
    t0 = a
    
    for j = 1:J 
        
        t[1] = t0 # t0 gets updated after every group J
        
        # Prediction loop
        
        y[1, 1, 1] = y0 # Initial condition for each group
        
        for m = 1:M
            t[m+1] = t[1] + m * h # time increment for each group
            y[1, m+1, 1]= y[1, m, 1]+ h* f(t[m], y[1, m, 1]) # Forward Euler update
        end
        
        # Correction loop
        
        for k =1:K
            
            y[1, 1, k+1]= y[1, 1, k]
            
            nfval= f(t, y[1, :, k]) # New function value 
            
            for m=1:M
                
                Q=Lagrange(t[1:k+1], nfval[1:k+1], t[m], t[m+1])  # Quadrature
                y[1, m+1, k+1] = y[1, m, k+1] + h*(f(t[m], y[1, m, k+1]) - f(t[m], y[1, m, k]))+Q
                
            end
        end
        
        y0=y[1, M+1, K+1] # storing initial value for the next group (Predictor)
        
        t0 = t[M+1] # updated initial time
        
        
    end
    
    return y 
    
end
```

```julia
# Convergence Test for IDC (Uniformly spaced, Forward Euler predictor and Corrector)

a = 0 # Initial time
b = 1 # Terminal time
y0 = 1 # Initial condition
p = 6 # desired order
M = p - 1 # Number of corrections

lambda = -1.0
Nk = 8

err = zeros(p, Nk)
ns = zeros(Nk)

for kk in 1:p
    for n in 1:Nk
        N = 20 * n
        ns[n] = N
        h = (b - a) / N
        #f(t, y)= -2* t .* y .^2
        f(t, y)= lambda* y
        #f(t, y)= y.^2
        ystore= idc6_FE(f, a, b, y0, h, N, p, M, kk - 1)

        #exact = 1.0 / 2
        exact = y0*exp(lambda * b)
        
        #exact=2.0
        
        err[kk, n] = abs(exact - ystore[1, end, kk])
    end
end
# Compute order of convergence and error plot
pdeg = polyfit(log10.(ns[:]), log10.(err[1, :]),1)
print("Order p= ",abs(pdeg[2]),"\n") 
plot(ns, err[1, :],xaxis=:log, yaxis=:log, label="Predictor")

for j in 2:p
    plot!(ns, err[j, :],xaxis=:log, yaxis=:log, label="$(j-1) Correction")
    pdeg = polyfit(log10.(ns[:]), log10.(err[j, :]),1)
    print("Order p= ", abs(pdeg[2]),"\n")
end
display(plot!(legend=:bottomleft))
```

#  IDC6: Predictor: RK-2, 1 Correction: RK-2, 2 Corrections: RK-2 

```julia
function idc6_Rk_2(f, a, b, y0, h, N, p, M, K)
    
    # Inputs: a: tspan[1], b: tspan[2], y0:initial cond, h=dt, N: # of intervals, 
    # Inputs: p= desired order, M : # of subintervals on each group, K: # of corrections,
    
    
    J = N ÷ M # Number of groups (÷ denotes the integer division)
    
   
    
    
    y=zeros(1, M+1, K+1)
    t = zeros(M+1)
    t0 = a
    
    for j = 1:J 
        
        t[1] = t0 # t0 gets updated after every group J
        
        # Prediction loop
        
        y[1, 1, 1] = y0 # Initial condition for each group
        
        for m = 1:M
            t[m+1] = t[1] + m * h # time increment for each group
            k1= h*f(t[m], y[1, m, 1])
            k2= h*f(t[m]+h, y[1, m, 1]+k1)
            y[1, m+1, 1]= y[1, m, 1]+ 0.5*(k1+k2) # RK
        end
        
        # Correction loop
        
        for k =1:K
            
            y[1, 1, k+1]= y[1, 1, k]
            
            nfval= f(t, y[1, :, k]) # New function value 
            
            for m=1:M
                
                yh=LagF(t[m:m+1], nfval[m:m+1], t[m]+h) # f(t_m+h, y(t_m+h))
                
                Q=Lagrange(t[1:(2*k+2)], nfval[1:(2*k+2)], t[m], t[m]+h)  # Quadrature
                
                k1= h*(f(t[m], y[1, m, k+1])-f(t[m], y[1, m, k]))
                
                #k2= h*(f(t[m]+h, y[1, m, k+1]+k1+Q)- f(t[m]+h, y[1, m+1, k]))
                
                k2= h*(f(t[m]+h, y[1, m, k+1]+k1+Q)- yh)
                
                y[1, m+1, k+1] = y[1, m, k+1] + 0.5*(k1+k2) +Q
                
            end
        end
        
        y0=y[1, M+1, K+1] # storing initial value for the next group (Predictor)
        
        t0 = t[M+1] # updated initial time
        
        
    end
    
    return y 
    
end
```

```julia
# Convergence Study for High-Order IDC (Uniformly spaced, RK-2 Predictor and Corrector)

a = 0 # Initial time
b = 1 # Terminal time
y0 = 1 # Initial condition
p = 6 # desired order
M = p - 1 # Number of corrections
Nk = 10

err = zeros(p, Nk)
ns = zeros(Nk)

cr=3 # Predictor (1) + # of corrections (2)

for kk in 1:cr
    for n in 1:Nk
        N = 20 * n
        ns[n] = N
        h = (b - a) / N
        f(t, y)= -2* t .* y .^2
        #f(t, y)= y
        #f(t, y)= y.^2
        ystore= idc6_Rk_2(f, a, b, y0, h, N, p, M, kk - 1)

        exact = 1.0 / 2
        #exact= exp(b)
        #exact=2.0
        err[kk, n] = abs(exact - ystore[1, end, kk])
    end
end
# Compute order of convergence and error plot
pdeg = polyfit(log10.(ns[:]), log10.(err[1, :]),1)
print("Order p= ",abs(pdeg[2]),"\n") 
plot(ns, err[1, :],xaxis=:log, yaxis=:log, label="Predictor", title="IDC6: RK-2 Predictor and 2 correction loops of RK-2")

for j in 2:cr
    plot!(ns, err[j, :],xaxis=:log, yaxis=:log, label="$(j-1) Correction")
    pdeg = polyfit(log10.(ns[:]), log10.(err[j, :]),1)
    print("Order p= ", abs(pdeg[2]),"\n")
end
display(plot!(legend=:bottomleft))
```

# IDC-6 Constructed with RK-3 (Predictor and one Corrector are computed using RK-3) method

```julia
function idc6_RK3(f, a, b, y0, h, N, p, M, K)
    
    # Inputs: a: tspan[1], b: tspan[2], y0:initial cond, h=dt, N: # of intervals, 
    # Inputs: p= desired order, M : # of subintervals on each group, K: # of corrections,
    
    # RK-3 parameters: 
    
    c2=1/2 
    c3=1 
    a21=1/2 
    a31= -1 
    a32=2
    
    b1=1/6  
    b2=2/3  
    b3=1/6
    
    # IDC
    
    J = N ÷ M # Number of groups (÷ denotes the integer division)
    
    
    y=zeros(1, M+1, K+1)
    t = zeros(M+1)
    t0 = a
    
    for j = 1:J 
        
        t[1] = t0 # t0 gets updated after every group J
        
        # Prediction loop
        
        y[1, 1, 1] = y0 # Initial condition for each group
        
        
        for m = 1:M
            t[m+1] = t[1] + m * h # time increment for each group
            
            k1= h*f(t[m], y[1, m, 1])
            
            k2= h*f(t[m]+ c2*h, y[1, m, 1] + a21*k1)
            
            k3=h*f(t[m]+ c3*h, y[1, m, 1] + a31* k1 + a32*k2)
            
            y[1, m+1, 1]= y[1, m, 1]+ (b1*k1+ b2*k2+ b3*k3) # RK-3 update
            
        end
        
        
        # Correction loop
        
        for k =1:K
            
            y[1, 1, k+1]= y[1, 1, k]
            
            nfval= f(t, y[1, :, k]) # New function value 
            
            for m=1:M
                
                # How to approximate f(t_m+h/2, y(t_m+h/2))?
                
                # Lagrange interpolation !! ? 
                
                yhalf=LagF(t[1:6], nfval[1:6], t[m]+ c2*h) # Approximating f(t_m+c2*h, y(t_m+c2*h))
                
                Q1=Lagrange(t[1:(4*k+2)], nfval[1:(4*k+2)], t[m], t[m]+ c2*h) # Aprroximating the integral
                
                Q2=Lagrange(t[1:(4*k+2)], nfval[1:(4*k+2)], t[m], t[m+1]) # Aprroximating the integral
                
                k1= h*(f(t[m], y[1, m, k+1])-f(t[m], y[1, m, k]))
                
                k2= h*(f(t[m]+ c2*h, y[1, m, k+1] + a21* k1 + Q1) -  yhalf) 
                
                k3= h*(f(t[m]+ c3*h, y[1, m, k+1] + a31* k1 + a32*k2 + Q2) - f(t[m]+ c3*h, y[1, m+1, k]))
                
                #k3= h*(f(t[m]+h, y[1, m, k+1] + 2*k2 - k1 + Q2)- LagF(t[m:m+1], nfval[m:m+1], t[m]+h))
                
                y[1, m+1, k+1] = y[1, m, k+1] + (b1*k1 + b2*k2 + b3*k3) + Q2
                
            end
        end
        
        y0=y[1, M+1, K+1] # storing initial value for the next group (Predictor)
        
        t0 = t[M+1] # updated initial time
        
        
    end
    
    return y 
    
end
```

```julia
# Convergence Test of IDC using High order intergrators (Uniformly spaced, RK-3 Predictor and Corrector)

a = 0 # Initial time
b = 1 # Terminal time
y0 = 1 # Initial condition
p = 6 # desired order
M = p - 1 # Number of corrections


Nk = 4

err = zeros(p, Nk)
ns = zeros(Nk)

cr=2 # Predictor (1) + # of corrections (1)

for kk in 1:cr
    for n in 1:Nk
        N = 5 * n
        ns[n] = N
        h = (b - a) / N
        #f(t, y)= -2* t .* y .^2
        f(t, y)= y
        #f(t, y)= y.^2
        
        ystore= idc6_RK3(f, a, b, y0, h, N, p, M, kk - 1)

        #exact = 1.0 / 2
        exact= exp(b)
        #exact=2.0
        err[kk, n] = abs(exact - ystore[1, end, kk])
    end
end
# Compute order of convergence and error plot
pdeg = polyfit(log10.(ns[:]), log10.(err[1, :]),1)
print("Order p= ",abs(pdeg[2]),"\n") 
plot(ns, err[1, :],xaxis=:log, yaxis=:log, label="Predictor")

for j in 2:cr
    plot!(ns, err[j, :],xaxis=:log, yaxis=:log, label="$(j-1) Correction", title="RK3 Predictor and one RK3 Correction")
    pdeg = polyfit(log10.(ns[:]), log10.(err[j, :]),1)
    print("Order p= ", abs(pdeg[2]),"\n")
end
display(plot!(legend=:bottomleft))
```

```julia
function idc7_FE_RK3(f, a, b, y0, h, N, p, M, K)
    
    # Inputs: a: tspan[1], b: tspan[2], y0:initial cond, h=dt, N: # of intervals, 
    # Inputs: p= desired order, M : # of subintervals on each group, K: # of corrections,
    
    # RK-3 parameters: 
    
    c2=1/2 
    c3=1 
    a21=1/2 
    a31= -1 
    a32=2
    
    b1=1/6  
    b2=2/3  
    b3=1/6
    
    # IDC
    
    J = N ÷ M # Number of groups (÷ denotes the integer division)
    
    
    y=zeros(1, M+1, K+1)
    t = zeros(M+1)
    t0 = a
    
    for j = 1:J 
        
        t[1] = t0 # t0 gets updated after every group J
        
        # Prediction loop
        
        y[1, 1, 1] = y0 # Initial condition for each group
        
        
        for m = 1:M
            t[m+1] = t[1] + m * h # time increment for each group
            
            t[m+1] = t[1] + m * h # time increment for each group
            y[1, m+1, 1]= y[1, m, 1]+ h* f(t[m], y[1, m, 1]) # Forward Euler update
            
        end
        
        
        # Correction loop
        
        for k =1:K
            
            y[1, 1, k+1]= y[1, 1, k]
            
            nfval= f(t, y[1, :, k]) # New function value 
            
            for m=1:M
                
                # How to approximate f(t_m+h/2, y(t_m+h/2))? 
                
                # Lagrange interpolation !!?
                
                
                yhalf=LagF(t[1:(3*k+1)], nfval[1:(3*k+1)], t[m]+ c2*h) # Approximating f(t_m+c2*h, y(t_m+c2*h))
                
                Q1=Lagrange(t[1:(3*k+1)], nfval[1:(3*k+1)], t[m], t[m]+ c2*h) # Aprroximating the integral
                
                Q2=Lagrange(t[1:(3*k+1)], nfval[1:(3*k+1)], t[m], t[m+1]) # Aprroximating the integral
                
                k1= h*(f(t[m], y[1, m, k+1])-f(t[m], y[1, m, k]))
                
                k2= h*(f(t[m]+ c2*h, y[1, m, k+1] + a21* k1 + Q1) -  yhalf) 
                
                k3= h*(f(t[m]+ c3*h, y[1, m, k+1] + a31* k1 + a32*k2 + Q2) - f(t[m]+ c3*h, y[1, m+1, k]))
                
                #k3= h*(f(t[m]+h, y[1, m, k+1] + 2*k2 - k1 + Q2)- LagF(t[m:m+1], nfval[m:m+1], t[m]+h))
                
                y[1, m+1, k+1] = y[1, m, k+1] + (b1*k1 + b2*k2 + b3*k3) + Q2
                
            end
        end
        
        y0=y[1, M+1, K+1] # storing initial value for the next group (Predictor)
        
        t0 = t[M+1] # updated initial time
        
        
    end
    
    return y 
    
end
```

# Multistep IDC

```julia
function idc_multistep(f, a, b, y0, h, N, p, M, K)
    
    # Inputs: a: tspan[1], b: tspan[2], y0:initial cond, h=dt, N: # of intervals, 
    # Inputs: p= desired order, M : # of subintervals on each group, K: # of corrections,
    
    
    J = N ÷ M # Number of groups (÷ denotes the integer division)
    
    
    y=zeros(1, M+1, K+1)
    
    t = zeros(M+1)
    
    t0 = a
    
    for j = 1:J 
        
        t[1] = t0 # t0 gets updated after every group J
        
        t[2]=t[1]+h
        
        # Prediction loop
        
        y[1, 1, 1] = y0 # Initial condition for each group
        
        # The starter is computed by the explicit RK-2 
        k1=h*f(t[1], y[1, 1, 1])
        k2=h*f(t[2], y[1, 1, 1]+k1)
        
        y[1, 2, 1]=y[1]+0.5*(k1+k2)
        
        
        
        
        for m = 2:M
            t[m+1] = t[1] + m * h # time increment for each group
            
            y[1, m+1, 1]= y[1, m, 1]+ (3/2)* h* f(t[m], y[1, m, 1]) - 0.5*h* f(t[m-1], y[1, m-1, 1]) # AB-2 update
        end
        
        # Correction loop
        
        for k =1:K
            
            y[1, 1, k+1]= y[1, 1, k]
            
            nfval= f(t, y[1, :, k]) # New function value 
            
            # First step (Starter) computed by Trapezoidal RK2 (Explicit)
            
            Qdr=Lagrange(t[1:(2*k+2)], nfval[1:(2*k+2)], t[1], t[2]) # Quadrature
            
            k1= h*(f(t[1], y[1, 1, k+1])- f(t[1], y[1, 1, k]))
            
            k2= h*(f(t[2], y[1, 1, k+1] + k1 + Qdr) - f(t[2], y[1, 2, k])) 
        
            y[1, 2, k+1]=y[1, 1, k+1] + k1/2 + k2/2 + Qdr
            
    
            
            
            for m=2:M
                
                Q= Lagrange(t[1:(2*k+2)], nfval[1:(2*k+2)], t[m], t[m+1])
                b1= (3/2)* h*(f(t[m], y[1, m, k+1]) - f(t[m], y[1, m, k]))
                b2= (1/2)* h*(f(t[m-1], y[1, m-1, k+1]) - f(t[m-1], y[1, m-1, k]))
                y[1, m+1, k+1]= y[1, m, k+1] + b1 - b2 + Q
                
            end
        end
       
        
        y0=y[1, M+1, K+1] # storing initial value for the next group (Predictor)
        
        t0 = t[M+1] # updated initial time
        
        
    end
    
    return y
    
end
```

```julia
# Convergence Study for Multi-step IDC (Uniformly spaced, AB-2 predictor and corrector)

a = 0 # Initial time
b = 1 # Terminal time
y0 = 1 # Initial condition
p = 6 # desired order
M = p - 1 # Number of corrections

Nk = 20
lambda = -1

ns = zeros(Nk)

cr=3 # Predictor (1) + # of corrections (2)

err = zeros(cr, Nk)

for kk in 1:cr
    
    for n in 1:Nk
        N = 10 * n
        
        ns[n] = N
        h = (b - a) / N
        
        f(t, y)= lambda* y
        #f(t, y)= -2* t .* y .^2
        
        ystore= idc_multistep(f, a, b, y0, h, N, p, M, kk - 1)

        
        exact=y0*exp(lambda*b)
        
        #exact= 1.0/2
        
        err[kk, n] = abs(exact - ystore[1, end, kk])
    end
end
# Compute order of convergence and error plot
pdeg = polyfit(log10.(ns[:]), log10.(err[1, :]),1)
print("Order p= ",abs(pdeg[2]),"\n") 
plot(ns, err[1, :],xaxis=:log, yaxis=:log, label="Predictor")

for j in 2:cr
    plot!(ns, err[j, :],xaxis=:log, yaxis=:log, label="$(j-1) Correction")
    pdeg = polyfit(log10.(ns[:]), log10.(err[j, :]),1)
    print("Order p= ", abs(pdeg[2]),"\n")
end
display(plot!(legend=:bottomleft))
```
