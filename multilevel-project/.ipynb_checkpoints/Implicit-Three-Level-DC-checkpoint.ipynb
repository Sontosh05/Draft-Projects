# Implicit Three Level Deferred Correction (Reference Solution)

```julia
function R3imp(f, a, b, u0, N)
    dt=(b-a)/N
    t=(0:N)*dt
    m=2
    DT1=m*dt
    DT2=m*DT1
    U=zeros(1, N+1, 3)
    
    U[1, 1, 1]=u0
    U[1, 1, 2]=u0
    U[1, 1, 3]=u0

    for i=1:m^2:N
        U[1, i, 2]=U[1, i, 3] # Injection from the third level
        for j=i:m:i+2
            U[1, j, 1]=U[1, j, 2] # Injection from the second level
            for k=j:j+1
                U[1, k+1, 1] = U[1, k, 1] / (1 .- dt)
            end
            U[1, j+2, 2]= (U[1, j, 2] .- DT1 * U[1, j+2, 1] + Lagrange(t[j:j+2], f(t[j:j+2], U[1, j:j+2, 1]), t[j], t[j+2])) / (1 .- DT1)
        end
        U[1, i+4, 3]= (U[1, i, 3] .- DT2 * U[1, i+4, 2] + Lagrange(t[i:m:i+4], f(t[i:m:i+4], U[1, i:m:i+4, 2]), t[i], t[i+4])) / (1 .- DT2)
    end
        
    return  U, t
end
```

```julia
# Convergence rate test of reference solution
using LinearAlgebra

u0 = 1.0
a=0
b=1.0
K = 4
err = zeros(K, 1)
ns=zeros(K, 1)
m=2
f(t, u) = u

for k in 1:K
    
    N = 8^k
    
    ns[k]=N
    Uim, t= R3imp(f, a, b, u0, N)
    exact= u0*exp.(t)
    err[k] = norm(exact[1:m^2:end] .- Uim[1, 1:m^2:end, 3], Inf)
end

# Compute order of convergence and error plot
pdeg = polyfit(log10.(ns[:]), log10.(err[:]),1)
print("convergence order = ",abs(pdeg[2]),"\n") 
plot(ns, err[:], xaxis=:log, yaxis=:log, label="3 Level Reference Solution")
display(plot!(legend=:topright))
```

```julia
function inter_imp(a, b, u0, dt, m)
    uk=zeros(m+1)
    uk[1]=u0
    tv=a:dt:b
    for i=1:m
        uk[i+1]=uk[i]/(1 .- dt)
    end
    return uk[end]
end
```

```julia
function Imp_num3(f, a, b, N, u0, H, K)
    
    dt = (b - a)/ N
    m = 2 
    DT1 = m*dt
    DT2 = m*DT1
    u = zeros(1, N + 1, K + 1)
    u[1, 1, 1] = u0
    ps = 3:m:N
    ps2=5:m^2:N
    u[1, ps2, 1] = rand(length(ps2)) # initializing random values
    t = (0:N)*dt
    up = zeros(N + 1)
    up[1] = u0
    uc = zeros(1, N + 1, K+1)
    uc[1, 1, 1] = u0
    uc[1, ps, 1] = rand(length(ps)) # initializing random values

    for k in 1:K
        
        u[1, 1, k+1] = u0
        uc[1, 1, k+1]= u0
        
        # F-relaxation on level 1
        for i=1:m:N
            up[i]=uc[1, i, k] # Injection from Level 2
            up[i+1]= H(t[i], t[i+1], up[i], dt, m-1) # One step solver
            #up[i+1]=up[i]/(1 .- dt)
        end
        # C-relaxation on level 1
        for j=2:2:N
            up[j+1]= H(t[j], t[j+1], up[j], dt, m-1)  # One step solver
            #up[j+1]= up[j]/(1 .- dt)
        end

        # Let's Continue Tomorrow...kids are getting ..
        
        # F-relaxation on level 2
        for n =1:m^2:N
            uc[1, n, k+1]=u[1, n, k] # Injection from the third level
            uc[1, n+2, k+1] = (uc[1, n, k+1] - DT1*up[n+2] + Lagrange(t[n:n+2], f(t[n:n+2], [uc[1, n, k]; up[n+1:n+2]]), t[n], t[n+2]))/(1 .- DT1)
        end
        # C-relaxation on level 2
        for l=3:m^2:N
            uc[1, l+2, k+1] = (uc[1, l, k+1] - DT1* f(t[l], uc[1, l+2, k]) + Lagrange(t[l:l+2], f(t[l:l+2], [uc[1, l, k]; up[l+1:l+2]]), t[l], t[l+2]))/(1 .- DT1)
        end

        # Coarse-grid correction on level 3
        for z=1:m^2:N
            uc[1, z, k+1]=u[1, z, k]
            u[1, z+4, k+1]=(u[1, z, k+1] - DT2*f(t[z], u[1, z+4, k])+Lagrange(t[z:m:z+4], f(t[z:m:z+4], uc[1, z:m:z+4, k+1]), t[z], t[z+4]))/(1 .- DT2)
        end
        
    end

    return  u, t
end
```

```julia
using LinearAlgebra, Plots

u0 = 1.0
a = 0
b = 1.0
#f(t, u) = t .* u.^2 .* cos.(t) .^2
f(t, u)= u
N = 16
K = 10 # Number of Iterations
m = 2
err=zeros(K+1)
u, t= Imp_num3(f, a, b, N, u0, inter_imp, K) # Approximate Solution
U, t = R3imp(f, a, b, u0, N) # Reference Solution
for k=1:K+1
    err[k]= norm( U[1, 1:m^2:end, 3] .- u[1, 1:m^2:end, k], Inf ) .+10^(-16)
end
plot(0:K, err, yaxis=:log, xlabel="Number of Iterations", ylabel="Error", label=" 3 Level Finite Step Convergence")
```

```julia
using SparseArrays

function Block_Ref3(N, dim, m)
    # Initialize an (N-1) x (N-1) sparse matrix
    dt = 1 / N
    dt1 = m * dt
    dt2 = m * dt1
    A = spzeros(dim, dim)

    # Set the first element separately
    A[1, 1] = 1

    alpha1 = 1 - dt1
    alpha2 = 1 - dt2
    beta = -1 + dt1 * ((5 / 6) * (1 / (1 - dt)^2) - (2 / 3) * (1 / (1 - dt)) - 1 / 6)

    # Fill the main diagonal and subdiagonal based on the correct pattern
    for i in 2:dim
        # Set 1 for specific indices and 2 for every third row (4th, 7th, 10th, etc.)
        if (i - 1) % 3 == 0  # Every third row (starting from row 4) gets 2
            A[i, i] = alpha2
        else
            A[i, i] = alpha1
        end

        # Set subdiagonal entries with 
        A[i, i - 1] = beta
    end

    # Add special rows with expressions involving dt2 at every 3rd row, starting from row 4
    for i in 4:3:dim
        A[i, i - 3] = -1 - dt2 / 6
        A[i, i - 2] = -2 * dt2 / 3
        A[i, i - 1] = (5 / 6) * dt2
    end

    # Convert to a dense matrix for display, if desired
    A_dense = Matrix(A)

    # Solve for u using A_dense and b
    u = zeros(dim)
    u[1] = 1.0
    b = zeros(dim)
    b[1] = 1.0
    u = A_dense \ b
    return u
end
```

```julia
using LinearAlgebra, Plots

u0 = 1.0
a = 0
b = 1.0
f(t, u)= u
N = 16
dim=N-3
m = 2
Uim, t= R3imp(f, a, b, u0, N)
u=Block_Ref3(N, dim, m)
norm(Uim[1, 1:4:end, 3] .- u[1:3:end])
```
