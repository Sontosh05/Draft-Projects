```julia
# Reference Solution (Implicit) for u'=lambda*u

function refimp(f, a, b, u0, N)
    dt=(b-a)/N
    t=(0:N)*dt
    m=2
    DT=m*dt
    U=zeros(1, N+1, 2)
    U[1, 1, 1]=u0
    U[1, 1, 2]=u0
    for i=1:m:N

        U[1, i, 1]=U[1, i, 2]
        
        for j=i:i+1
            U[1, j+1, 1]= U[1, j, 1]/(1-dt*lambda)
        end
           
        U[1, i+m, 2]= (U[1, i, 2]- DT*lambda*U[1, i+m, 1] + Lagrange(t[i:i+2], f(t[i:i+2], U[1, i:i+2, 1]), t[i], t[i+2]))/(1-DT*lambda)

        #U[1, i+m, 2]= (U[1, i, 2]- DT*lambda*U[1, i+m, 1] + (DT/6)*lambda*U[1, i, 1]+(2*DT/3)*lambda*U[1, i+1, 1]+(DT/6)*lambda*U[1, i+m, 1])/(1-DT*lambda)
        
        # Trapezoidal
        #U[1, i+m, 2]= U[1, i, 2]+ DT*(f(t[i], U[1, i, 2])-f(t[i], U[1, i, 1])) + DT*0.5*(f(t[i], U[1, i, 1])+f(t[i+2], U[1, i+2, 1]))
    end
        
    return  U, t
end
```

```julia
# Convergence test
using LinearAlgebra

u0 = 1.0
lambda= 1
a=0
b=1.0
K = 6
err = zeros(K, 1)
ns=zeros(K, 1)
m=2
f(t, u) = lambda * u

for k in 1:K
    
    N = 8*2^k
    
    ns[k]=N
    U, t= refimp(f, a, b, u0, N)
    exact= u0*exp.(lambda .* t)
    err[k] = norm(exact[1:m:end] .- U[1, 1:m:end, 2], Inf)
end

# Compute order of convergence and error plot
pdeg = polyfit(log10.(ns[:]), log10.(err[:]),1)
print("convergence order = ",abs(pdeg[2]),"\n") 
plot(ns, err[:], xaxis=:log, yaxis=:log, label="Reference Solution")
display(plot!(legend=:topright))
```

```julia
function inter(a, b, u0, dt, m)
    uk=zeros(m+1)
    uk[1]=u0
    tv=a:dt:b
    lambda=1.0
    for i=1:m
        uk[i+1]=uk[i]/(1-lambda*dt)
    end
    return uk[end]
end
```

```julia
function twolevel(f, a, b, N, u0, m, K)
    dt = (b - a)/ N 
    DT = m*dt
    u = zeros(1, N + 1, K + 1)
    u[1, 1, 1] = u0
    ps = 3:m:N
    u[1, ps, 1] = ones(length(ps)) # initializing random values
    t = (0:N)*dt
    up = zeros(N + 1)
    up[1] = u0

    for k in 1:K
        
        u[1, 1, k + 1] = u0

        #coarsening factor 4

        # F-Relaxation
        for i=1:m:N
            up[i]=u[1, i, k] # injection from the second level
            up[i+1]=up[i]/(1-dt*lambda)
        end

        # C-Relaxation
        for j=2:m:N
            up[j+1]=up[j]/(1-dt*lambda)
        end

        # Coarse Grid Correction
        for n =1:m:N
            up[n]=u[1, n, k] 
            
            u[1, n+m, k+1] = (u[1, n, k+1] - DT*lambda*up[n+m] + Lagrange(t[n:n+m], f(t[n:n+m], up[n:n+m]), t[n], t[n+m]))/(1-lambda*DT)

            # Trapezoidal
            #u[1, n+m, k+1] = u[1, n, k+1] + DT*(f(t[n], u[1, n, k+1]) - f(t[n], u[1, n, k])) + DT*0.5*(f(t[n], up[n])+f(t[n+m], up[n+m]))
        end
    end

    return t, u
end
```

```julia
using LinearAlgebra, Plots

u0 = 1.0
a = 0
b = 1.0
lambda=1.0
f(t, u) = lambda*u
N = 60
K = 10 # Number of Iterations
m = 2
err=zeros(K+1)
tval, u= twolevel(f, a, b, N, u0, m, K) # Approximate Solution
U, t = refimp(f, a, b, u0, N) # Reference Solution
for k=1:K+1
    err[k]= norm(U[1, 1:m:end, 2] .- u[1, 1:m:end, k], Inf) .+10^(-16)
end
plot(0:K, err, yaxis=:log, xlabel="Number of Iterations", ylabel="Error")
```

```julia
# RIDC Implicit
function ridcimp(f, tspan, u0, dt, N, K)
    lambda=1.0
    t=(0:N)*dt
    u=zeros(1, N+1, K+1)
    u[1, 1, 1]=u0

    #Predictor
    for i=1:N
        u[1, i+1, 1]= [1, i, 1]/(1-dt*lambda)
    end

    # Corrector
    for k=1:K
        u[1, 1, k+1]=u0
        for j=1:N
            u[1, j+1, k+1]=u[1, j, k+1]-dt*lambda*u[1, j+1, k]+ Lagrange(t[j:j+1], f(t[j:j+1], u[1, j:j+1, k]), t[j], t[j+1])
        end
    end
    return u, t
end  
```

```julia
# Test #2
using Plots
M = 6
N_values = zeros(M)

for i in 1:M
    N_values[i] = 50* i
end

lambda=1.0
f(t, u)= lambda*u
tspan=[0 1.0]
u0=1.0
K = 1
err1 = zeros(M, K + 1)

for p in 1:M
   
    N = convert(Int, N_values[p])
    dt = (tspan[2] - tspan[1]) / N

    (u, t) = ridcimp(f, tspan, u0, dt, N, K)
    

    exact1=exp.(lambda*t)
    # Storing errors
    for i in 1:K+1
        err1[p, i] = norm(exact1[N+1] .- u[1, N+1, i])
    end
end

# Compute order of convergence and error plot
pdeg = polyfit(log10.(N_values[:]), log10.(err1[:, 1]),1)
print("convergence order = ",abs(pdeg[2]),"\n") 
plot(N_values, err1[:, 1], xaxis=:log, yaxis=:log, label="Provisonal")

for j in 2:K+1
    plot!(N_values, err1[:,j],xaxis=:log, yaxis=:log, label="RIDC-$(j-1)")
    pdeg = polyfit(log10.(N_values[:]), log10.(err1[:, j]),1)
    print("convergence order = ", abs(pdeg[2]),"\n")
end
display(plot!(legend=:topright))
```
