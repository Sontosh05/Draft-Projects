```julia
using Plots
include("polyfit_function.jl")
include("Lagrange_interpolation.jl")
include("LReg.jl")
```

```julia
function Ref(f, a, b, u0, N)
    dt=(b-a)/N
    t=(0:N)*dt
    m=2
    DT=m*dt
    U=zeros(1, N+1, 2)
    U[1, 1, 1]=u0
    U[1, 1, 2]=u0
    
    for i=1:m:N

        U[1, i, 1]=U[1, i, 2]
        for j=i:i+1
            U[1, j+1, 1]= U[1, j, 1]+dt*f(t[j], U[1, j, 1])
        end
           
        U[1, i+2, 2]= U[1, i, 2]+ DT*(f(t[i], U[1, i, 2])-f(t[i], U[1, i, 1])) + LReg(t[i:i+2], f(t[i:i+2], U[1, i:i+2, 1]), t[i], t[i+2])
    end
        
    return  U, t
end
```

```julia
function inter(a, b, u0, dt, m)
    uk=zeros(m+1)
    uk[1]=u0
    tv=a:dt:b
    for i=1:m
        uk[i+1]=uk[i]+dt* f(tv[i], uk[i])
    end
    return uk[end]
end
```

```julia
function twolevel(f, a, b, N, u0, H, K)
    dt = (b - a)/ N
    m = 2 
    DT = m*dt
    u = zeros(1, N + 1, K + 1)
    u[1, 1, 1] = u0
    ps = 3:m:N
    u[1, ps, 1] = ones(length(ps)) # initializing random values
    t = (0:N)*dt
    up = zeros(N + 1)
    up[1] = u0

    for k in 1:K
        
        u[1, 1, k + 1] = u0

        for i=1:m:N
            up[i]=u[1, i, k]
            up[i+1]=H(t[i], t[i+1], up[i], dt, m-1)
        end
        for j=2:2:N
            up[j+1]=H(t[j], t[j+1], up[j], dt, m-1)
        end

        for n =1:m:N
            up[n]=u[1, n, k] # injected value
            u[1, n+2, k+1] = u[1, n, k+1] + DT*(f(t[n], u[1, n, k+1]) - f(t[n], u[1, n, k])) + LReg(t[n:n+2], f(t[n:n+2], up[n:n+2]), t[n], t[n+2])
        end
    end

    return t, u
end
```

```julia
using LinearAlgebra, Plots

u0 = 1.0
a = 0
b = 1.0
f(t, u) = -u .+ sin.(10*t)
N = 50
K = 10 # Number of Iterations
m = 2
err=zeros(K+1)
tval, u= twolevel(f, a, b, N, u0, inter, K) # Approximate Solution
U, t = Ref(f, a, b, u0, N) # Reference Solution
for k=1:K+1
    err[k]= norm(U[1, 1:m:end, 2] .- u[1, 1:m:end, k], Inf) .+10^(-16)
end
plot(0:K, err, yaxis=:log, xlabel="Number of Iterations", ylabel="Error")
```
