```julia
using Plots
include("polyfit_function.jl")
include("Lagrange_interpolation.jl")
include("LReg.jl")
```

```julia
function reference(f, a, b, U0, N, cr)
    
    m= 2 # coarsenning factor
    
    J= NÃ·m # number of time steps in coarse level
    
    dt=(b-a)/N
    DT=m*dt
    
    # Reference solution
    U=zeros(1, m+1, 2)
    t=zeros(m+1)
    t0=a

    for j=1:J
        
        t[1]=t0
        U[1, 1, 1]=U0
        for i= 1:m
            t[i+1]=t[1]+i*dt
            U[1, i+1, 1]= U[1, i, 1] + dt* f(t[i], U[1, i, 1])
        end
        
        # Correction loop
        for k=1:cr
            U[1, 1, k+1]= U[1, 1, k]
            for i=2:m
                U[1, i+1, k+1]= U[1, i-1, k+1]+ DT*(f(t[i-1], U[1, i-1, k+1])- f(t[i-1], U[1, i-1, k]))+ LReg(t[1:m+1], f(t[1:m+1], U[1, 1:m+1, k]), t[1], t[m+1])
            end
        end
        U0= U[1, m+1, cr+1]
        t0=t[m+1]    
    end
    return U
end
```

```julia
# Order of Convergence for reference solution

a=0
b=1
U0 = 1.0 # Initial condition

Nk = 8
p=2
err = zeros(Nk)
ns = zeros(Nk)


    for n in 1:Nk
        N =10* n
        ns[n] = N

        f(t, u)=  -2* t .* u .^2
        
        
        ustore= reference(f, a, b, u0, N, 1)

        exact = 1.0 / 2
        
        
        err[n] = abs(exact - ustore[1, end, 2])
    
    end
   

# Compute order of convergence and error plot
pdeg = polyfit(log10.(ns[:]), log10.(err[:]),1)
print("Order p= ",abs(pdeg[2]),"\n") 
plot(ns, err[:],xaxis=:log, yaxis=:log, label="Rate of convergence of reference solution")
```

```julia
function grid2c(f, a, b, u0, N, K)
    dt=(b-a)/N
    t=(0:N)*dt
    m=2
    DT=m*dt
    u=zeros(1, N+1, K+1)
    u[1, 1, 1]=u0
    ps=3:m:N
    u[1, ps, 1] = rand(length(ps))


    # 2 level iteration
    #fk=zeros(m+1)
    for k=1:K
        u[1, 1, k+1]=u0
        for i=1:m:N
            
            # Sequential solution on each time slice
            fk1=u[1, i, k]
            fk2=u[1, i, k]+dt*f(t[i], u[1, i, k])
            fk3=fk2+dt*f(t[i+1], fk2)
            # Updated solution on the coarse nodes
            u[1, i+2, k+1]= u[1, i, k+1]+ DT*(f(t[i], u[1, i, k+1])-f(t[i], fk1)) + LReg(t[i:i+2], f(t[i:i+2], [fk1, fk2, fk3]), t[i], t[i+2])
        end
    end
        
    return t, u
end
```

```julia
# Convergence test
using LinearAlgebra

u0 = 1.0
a=0
b=10

nF=2

cr=1

K = 20 # Number of Iterations

err = zeros(K+1)

f(t, u) = - u .+ sin.(10*t)
 
N = 50
    
t, u=grid2c(f, a, b, u0, N, K)
U=reference(f, a, b, u0, N, cr)
    
   
for k=1:K+1
    err[k] = norm(U[1, end, 2] - u[1, end, k])
end
using Plots
plot(0:K, err, yaxis=:log)
```
